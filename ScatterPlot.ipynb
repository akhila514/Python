```python
df1 = pd.DataFrame({'x':[1,2,3,4,5],'y':[4,8,16,20,24],'group':['A','B','C','D','None']})
sns.regplot(data = df1,x= "x", y = "y", marker= "+")
```

```python
p1 = sns.regplot(data = df1,x= "x", y = "y", marker= "o")
for line in range(0,df1.shape[0]):
    p1.text(df1.x[line]+0.2, df1.y[line], df1.group[line], horizontalalignment = 'left')
```
