```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#basic
x = np.arange(0,20)
y = np.arange(10,30)
```

```python
x
```

```python
y
```

```python
#plotting
#scatter
plt.scatter(x,y,c="r")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Basic 2D plot")
plt.savefig("Basic2DPlot.png")
```

```python
z = x*y
```

```python
plt.plot(x,z,'b-*',linewidth =2,linestyle='dashed',markersize=12)
```

```python
#subplots
```

```python
plt.subplot(4,4,1)
plt.plot(x,y,'r--')
```

```python
plt.subplot(4,4,2)
plt.plot(y,z,'g')
```

```python
plt.subplot(4,4,1)
plt.plot(x,y,'r--')

plt.subplot(4,4,2)
plt.plot(y,z,'g')

plt.subplot(4,4,3)
plt.plot(x,y,'b*-')

plt.subplot(4,4,4)
plt.plot(y,z,"ro")

plt.subplot(4,4,5)
plt.plot(y,z,"go")
```

```python
#simple equations
x =np.arange(10,20)
y = 3*x+1
plt.plot(x,y)
```

```python
#sine cosine
np.pi
```

```python
x = np.arange(0, 10 * np.pi, 0.1) 
y = np.sin(x) 
z = np.cos(x)
plt.plot(x,y,'ro')
plt.plot(x,z,'go')
```

```python
#bar plot
data = [[10, 25, 70, 20],
[40, 33, 51, 17],
[15, 42, 45, 29]]
X = np.arange(4)
plt.bar(X + 0.00, data[0], color = 'b', width = 0.25)
plt.bar(X + 0.25, data[1], color = 'g', width = 0.25)
plt.legend(labels=['Men', 'Women'])
```

```python
#histogram
h = np.array([5,20,10,55,60,12,15,2,70,33,45,44,42,43,46,47,45])
plt.hist(h)
plt.title("histogram")
```

```python
h = np.array([5,20,10,55,60,12,15,2,70,33,45,44,42,43,46,47,45])
plt.hist(h,bins=40)
plt.title("histogram")
```

```python
#boxplt
data = [np.random.normal(0, std, 50) for std in range(1, 5)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False);  
```

```python
#pie chart
```

```python
labels = 'laptop', 'Mobile', 'ipod', 'tab'
sizes = [100, 500, 20, 60]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0.2, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.2f%%', shadow=True)

plt.axis('equal')
plt.show()
```
