```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
#distplot
sns.distplot(df['PetalLengthCm'])
```

```python
#bins
sns.distplot(df['PetalLengthCm'],bins=15,kde=False)
```

```python
#kde
sns.distplot(df['PetalLengthCm'],kde=False,hist=True)
```

```python
#kde,rug
sns.distplot(df['PetalLengthCm'],kde=True,hist=True,rug=True)
```

```python
#rug custom
sns.distplot(df['PetalLengthCm'],kde=True,hist=True,rug=True,rug_kws={"color":"r","alpha":0.3,"linewidth":2,"height":0.1})
```

```python
#kde custom
sns.distplot(df['PetalLengthCm'],kde=True,hist=True,rug=True,kde_kws={"color":"g","alpha":0.5,"linewidth":5,"shade":True})
```

```python
#vertical hist
sns.distplot(df['SepalLengthCm'],vertical=True)
```

```python
sns.distplot(df['SepalLengthCm'])
```

```python
#histogram with a boxplot on top
#cut the window into 2 parts
f,(axis_box,axis_hist) = plt.subplots(2,sharex=True,gridspec_kw={"height_ratios":(.20,.80)})
sns.boxplot(df['SepalLengthCm'],ax=axis_box)
sns.distplot(df['SepalLengthCm'],ax=axis_hist)
```

```python
#with serveral variables
sns.distplot(df['SepalLengthCm'],color='blue',label='Sepal Length')
sns.distplot(df['SepalWidthCm'],color='red',label='Sepal Width')

plt.legend()
```

```python
f,axes = plt.subplots(2,2,figsize=(9,9),sharex=True)
sns.distplot(df['SepalLengthCm'],color='lightblue',ax=axes[0,0])
sns.distplot(df['SepalWidthCm'],color='lightgreen',ax=axes[0,1])
sns.distplot(df['PetalLengthCm'],color='purple',ax=axes[1,0])
sns.distplot(df['PetalWidthCm'],color='orange',ax=axes[1,1])
```
